Can you create a graph that shows the ED admission rate for each county, plotting under 18 against 18 and over, displaying separately each sex?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_asthma = pd.read_csv("../datasets/asthma.csv")
```

```python
df_asthma.head()
```

```python
df_asthma['Strata Name'].unique()
```

```python
df_asthma['Strata'].unique()
```

```python
#creating a new dataframe where strata is equal to sex. Leaves out Total Pop and race-ethnicity
df1= df_asthma[df_asthma['Strata'] == 'Sex']
#filtering the new dataframe that was created to include only age groups 18 and over and under 18
df1 = df1[(df1['Age Group'] == '18 and Over') | (df1['Age Group'] == 'Under 18')]
#first 5 lines of the new dataframe
df1.head()
```

```python
fig, ax = fig, ax = plt.subplots(figsize=(13,10))
sns.set(rc = {'figure.figsize':(8,8),"figure.dpi":300})
# This one will group all the sex together instead of sep. by sex 
ax = sns.boxplot(data=df1,
                 x='Age Group', y='Rate', hue='Strata',
                )
ax.set_ylim([0,210])
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

```python
fig, ax = fig, ax = plt.subplots(figsize=(13,10))
sns.set(rc = {'figure.figsize':(8,8),"figure.dpi":300})
# This one will sep. by sex
ax = sns.boxplot(data=df1,
                 x='Age Group', y='Rate', hue='Strata Name',
                )
ax.set_ylim([0,210])
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

# W3_HW11

Can you make the following boxplot look similar to the one we created using the df_pivot dataframe but instead, by using the original dataframe (df_asthma)? Hint: You will need to investigate your data, and may need to weed out the outliers.

    df_asthma.plot(x='Year', y='Rate', kind='box')
    plt.show()

```python
import pandas as pd
import matplotlib.pyplot as plt
df_asthma = pd.read_csv("../datasets/asthma.csv")
```

```python
df_asthma.head()
```

```python
df_asthma.describe()
```

```python
df_asthma.plot(x='Year', y='Rate', kind='box')
plt.show()
#this gives us a boxplot with only one column
```

```python
import seaborn as sns
#seaborn will allow us to make a boxplot with the original dataframe 
bxplt = sns.boxplot(data=df_asthma,
                 x='Year', y="Rate"
                )
plt.ylim(0,1000)
#this creates a boxplot that still contains a lot of outliers and isn't very helpful
plt.show()
```

```python
bxplt = sns.boxplot(data=df_asthma,
                 x='Year', y="Rate"
                )
plt.ylim(0,200)
plt.show()
#this boxplot still contains a lot of outliers 
```

```python
#adding in fliersize=0 means that the outliers won't be shown on the graph
bxplt = sns.boxplot(data=df_asthma,
                 x='Year', y="Rate", fliersize= 0
                )
#changing the range to show the boxplots a bit clearer
plt.ylim(0,150)
plt.show()
```

```python
# another possible way to extract out the outliers 
# get the IQR and outliers

Q1 = df_asthma['Rate'].quantile(0.25)
Q3 = df_asthma['Rate'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df = df_asthma[(df_asthma['Rate'] >= lower_bound) & (df_asthma['Rate'] <= upper_bound)]
```

```python
# this boxplot will show us the data within our defined IQR
sns.boxplot(y=df['Rate'], x=df_asthma['Year'], fliersize=1);
plt.show()
```
