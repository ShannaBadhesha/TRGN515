```python
#Code to get the reverse output of a DNA sequence

def reverse(DNAseq):
    return(DNAseq[::-1])

reverse(DNAseq)
```

```python
#Code to get the complement output of a DNA sequence

def complement(DNAseq):
    comp= ''
    DNAseq = DNAseq.upper()
    for letter in DNAseq:
        if letter == 'A':
            comp += 'T'
        elif letter == 'C':
            comp += 'G'
        elif letter == 'T':
            comp += 'A'
        elif letter == 'G':
            comp += 'C'
    return comp

complement(DNAseq)
```

```python
#Code to get the reverse complement output of a DNA sequence

def revcomp(DNAseq):
    comp= ''
    DNAseq = DNAseq.upper()
    for letter in DNAseq:
        if letter == 'A':
            comp += 'T'
        elif letter == 'C':
            comp += 'G'
        elif letter == 'T':
            comp += 'A'
        elif letter == 'G':
            comp += 'C'
    return comp[::-1]

revcomp(DNAseq)
```

#### Provide explanations for what complement and reverse complement sequences mean in molecular biology.

Complement strands are the opposite nucleotides of the original sequence.  A is the complement of T, T is the complement of A, G is the complement of C, and C is the complement of G. Because of the complements, it is able to bind together and both strands are going in the same direction. The complementary strand helps create a new strand by using the complements for each of the nucleotides. For complement in a double stranded helix, it would be going in the same direction and held together by hydrogen bonds. Reverse complement (antiparallel DNA) can help find the order of nucleotides and find complementary parts to RNA. This is needed for understanding gene expression in molecular biology. Reverse complement is the exact opposite order of the original strand.
